```python
import requests
import csv
import random
import time

api_key = 'RGAPI-e9f1332b-da10-4bf0-aef8-d87b130e32cf'
region = 'euw1'
routing_value = 'europe'
tier = 'SILVER'
queue = 'RANKED_SOLO_5x5'
headers = {'X-Riot-Token': api_key}

def get_random_summoners(count=500):
    url = f'https://{region}.api.riotgames.com/lol/league/v4/entries/{queue}/{tier}/I'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        summoners_data = response.json()
        if len(summoners_data) > count:
            return random.sample(summoners_data, k=count)
        return summoners_data
    return []

def get_puuid_from_summoner_id(summoner_id):
    url = f'https://{region}.api.riotgames.com/lol/summoner/v4/summoners/{summoner_id}'
    response = requests.get(url, headers=headers)
    time.sleep(0.7)  # Rate limiting
    if response.status_code == 200:
        data = response.json()
        return data.get('puuid')
    return None

def fetch_and_save_matches(puuid, count=5, desired_patch='14.8.576.8838'):
    url = f'https://{routing_value}.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count={count}'
    response = requests.get(url, headers=headers)
    time.sleep(0.7)  # Rate limiting
    matches_data = []
    if response.status_code == 200:
        match_ids = response.json()
        for match_id in match_ids:
            detail_url = f'https://{routing_value}.api.riotgames.com/lol/match/v5/matches/{match_id}'
            match_response = requests.get(detail_url, headers=headers)
            time.sleep(0.7)  # Rate limiting
            if match_response.status_code == 200:
                match_details = match_response.json()
                game_version = match_details['info']['gameVersion']
                if game_version == desired_patch:
                    match_row = [match_id, game_version]
                    win_red = 0
                    red_team, blue_team = [], []
                    for participant in match_details['info']['participants']:
                        if participant['teamId'] == 100:
                            red_team.append(participant['championName'])
                            if participant['win']:
                                win_red = 1
                        else:
                            blue_team.append(participant['championName'])
                    match_row.extend(red_team + blue_team)
                    match_row.append(win_red)
                    matches_data.append(match_row)
    return matches_data

def save_to_csv(matches_data, filename='lang.csv'):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['gameId', 'patchVersion', 'champion1', 'champion2', 'champion3', 'champion4', 'champion5',
                         'champion6', 'champion7', 'champion8', 'champion9', 'champion10', 'win_red'])
        writer.writerows(matches_data)

def main():
    start_time = time.time()
    summoners = get_random_summoners()
    all_matches_data = []
    for summoner in summoners:
        puuid = get_puuid_from_summoner_id(summoner['summonerId'])
        if puuid:
            matches_data = fetch_and_save_matches(puuid)
            all_matches_data.extend(matches_data)
    save_to_csv(all_matches_data)
    print("Test 1 (50 players, 5 game each) took:", time.time() - start_time, "seconds")

if __name__ == "__main__":
    main()
```
